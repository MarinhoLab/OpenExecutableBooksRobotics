# Valid imports

```python
from math import pi, sin, cos
import numpy as np
```

# Exercises

## Exercise a

```python
θ_a = pi/4.0

R_a =  np.array([[cos(θ_a),-sin(θ_a)],
                 [sin(θ_a), cos(θ_a)]])

# Printing the result is NOT a mandatory part of the answer.
print(f'R_a = {R_a}')
```

## Exercise b

```python
θ_b1 = pi/12.0
θ_b2 = -pi/2.0

R_b1 =  np.array([[cos(θ_b1),-sin(θ_b1)],
                  [sin(θ_b1), cos(θ_b1)]])

R_b2 =  np.array([[cos(θ_b2),-sin(θ_b2)],
                  [sin(θ_b2), cos(θ_b2)]])

R_b = R_b1 @ R_b2

print(f'R_b = {R_b}')
```

## Exercise c

```python
θ_c = pi/3.0 # As given in the exercise
x_c = 2.0 # As given in the exercise
y_c = 5.0 # As given in the exercise

H_c1 = np.array([[cos(θ_c),-sin(θ_c), 0],
                 [sin(θ_c), cos(θ_c), 0],
                 [0,        0,        1]])

H_c2 = np.array([[1,0,x_c],
                 [0,1,y_c],
                 [0,0,1]])

H_c = H_c1 @ H_c2

print(f'H_c = {H_c}')
```
